# Dividing Data set 

```python
df['Quality']=0
df.loc[df['quality']>6, 'Quality']=1
df.head()
```

```python
sns.countplot(df['Quality'])
```

# Over sampling 

```python
from sklearn.utils import resample, shuffle
```

```python
# The data is still skewed so we can use sklearn resample to generate more data from exisitng data
zeros = df[df['Quality'] == 0]
ones = df[df['Quality'] == 1]

unsampled = resample(ones, replace = True, n_samples = zeros.shape[0])
df = pd.concat([zeros, unsampled])
df = shuffle(df)
```

```python
plt.figure(figsize = (8,6))
sns.countplot(df['Quality'])
plt.title('Quality after resampling')
plt.show()
```

```python
# sns.pairplot(df, hue = 'Quality')
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn import preprocessing
```

```python
X = df.drop('Quality', axis = 1)
y = df['Quality']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)

ss=StandardScaler()

X_train= ss.fit_transform(X_train.values)
X_test= ss.transform(X_test.values)
X_train= pd.DataFrame(X_train, columns= X.columns)
X_test=pd.DataFrame(X_test, columns=X.columns)
X_train.head()
```

```python
X_test.head()
```

```python
plt.figure(figsize = (12,8))
sns.heatmap(df.corr(), annot = True, cmap = 'coolwarm')
plt.show()
```

# Models Traning 

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from xgboost import XGBClassifier
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score
from sklearn.metrics import precision_score, recall_score, f1_score, roc_auc_score
from sklearn.metrics import auc, roc_curve, confusion_matrix, classification_report, accuracy_score
```

# Random Forest Classifier 

```python
model = RandomForestClassifier()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
y_probs = model.predict_proba(X_test)[:,1]

print(precision_score(y_test, y_pred))
print(recall_score(y_test, y_pred))
print(f1_score(y_test, y_pred))
print(roc_auc_score(y_test, y_pred))
```

```python
print(classification_report(y_test, y_pred))
```

```python

print(f'ROC AUC SCORE: {roc_auc_score(y_test, y_probs)}')
```

```python
print('Cross Validation score with 5 folds')
print(f'ROC AUC {round(cross_val_score(model, X, y, cv = 5, scoring = "roc_auc").mean(),2)}')
print(f'Precision {round(cross_val_score(model, X, y, cv = 5, scoring = "precision").mean(),2)}')
print(f"F1 {round(cross_val_score(model, X, y, cv = 5, scoring = 'f1').mean(),2)}")
print(f'Recall {round(cross_val_score(model, X ,y, cv = 5, scoring = "recall").mean(),2)}')
```

```python
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot = True, fmt = 'd', cmap = 'coolwarm')
plt.show()
```

```python
fpr, tpr, threshold = roc_curve(y_test, y_probs)
roc_auc = auc(fpr, tpr)
sns.set_theme(style = 'white')
plt.figure(figsize=(8,6))
plt.plot(fpr, tpr, color = 'blue', label = 'AUC = %0.3f'% roc_auc)
plt.plot([0,1],[0,1], linestyle = '--', color = 'green')
plt.xlabel('False positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
plt.show()
```

```python
accuracy_score(y_test, y_pred)
```
